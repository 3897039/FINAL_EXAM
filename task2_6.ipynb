4)  Вы решили сравнивать метрику CPA в двух группах. Размер выборки - 2350 элементов в каждой группе. 
Для проверки нормальности распределения на выборке в 2350 наблюдений применили , 
критерий Шапиро-Уилка и получили p-value, равный 0.00002, alpha = 5%. 
Какой бы вывод мы могли сделать в данном случае?

Полученный p-value значительно меньше уровня значимости alpha, что говорит о статистической значимости различий и отклонении от нормального распределения в данной выборке.  Из этого можно сделать вывод, что данные распределены ненормально и применять параметрические методы статистики нецелесообразно. Поэтому лучше применять непараметрические критерии. 

В этом случае какой статистический критерий для проверки первоначальной гипотезы 
тут лучше всего подойдёт и почему?

Наиболее подходящим критерием является критерий Манна-Уитни, который является 
непараметрическим аналогом критерия Стьюдента для проверки различий между двумя независимыми выборками. 
Этот критерий не требует нормального распределения и меньше чувствителен 
к изменениям и отклонениям в распределении.  

5) Мы провели АБ-тест на увеличение average timespent per user. 
По итогам тестирования мы получили следующие данные. 
Является ли результат статистически значимым с уровнем доверия 80%? 
Какую версию мы выкатим на продакшн? 


A) Средняя - 360, отклонение - 40, количество - 9802 
B) Средняя - 352, отклонение - 58, количество - 9789


Для выявления статистической значимости А/В тест подходит  t-тест для сравнения двух средних значений. 
Основная гипотеза G0- средние значения времени пользователями в версиях А  и В равны. 
Вспомогательная гипотеза G1 – средние значения времени пользователями в версиях А  и В различны.

Вычислим стандартную ошибку для каждой версии:
SE_A = 40/ sqrt(9802) = 0.404 – для версии А
SE_A = 580/ sqrt(9789) = 0,585 – для версии В
Вычисление t-статистики:
t = (сред A – сред B) / sqrst(SE_A^2 + SE_B^2)=(3660-352)/ sqrst(0.404^2 + 0.585^2) = 2.091
По таблице t-распределения критическое значение t-статистики в итоге равно 1,282.


Так как вычисленное превышает статистическое значение, то можно отвергнуть нулевую гипотезу о равенстве средних и принять гипотезу о различии средних на уровне доверия 80%.
В итоге получается, что разница в среднем проведенном времени у версий А и В  является статистически значимой.
В итоге, для увеличения average timespent per user нужно использовать версию А как более продуктивную для достижения успеха.
